```python
import pandas as pd
```

```python
main = pd.read_csv('info.csv')
```

```python
main
```

```python
for i in range(40):
    temp = main.loc[i]
    new_dataframe = pd.DataFrame(temp).T
    new_dataframe['key']=0;
    data_x = pd.read_csv('data/patient_'+str(i)+'.csv', header=None).drop([0],axis=1)
    data_x['key']=0;
    merged_df = pd.merge(new_dataframe,data_x).drop(['key'],axis=1)
    merged_df.to_csv('data/merged_patient'+str(i)+'.csv')
```
